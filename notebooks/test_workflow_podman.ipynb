# Test the workflow with podman-compose

copy `.env_template` to `.env`. Never use the same passwords in production! But for testing here we use the default ones.

```bash
podman-compose -f docker-compose.db_neo.yml up -d
```

**Tipp:** Before running the following command, please close all other applications that might use significant system resources, as running containers can be resource-intensive. 

````python
engine = create_engine("mysql+pymysql://biokb_user:biokb_password@localhost:3306/biokb")

DbManager(engine).import_data()
TurtleCreator(engine).create_ttls()
````

Open Neo4j [browser](http://localhost:7474), login with neo4j/neo4j_password

````python
Neo4jImporter(
    neo4j_uri="bolt://localhost:7687", neo4j_user="neo4j", neo4j_pwd="neo4j_password"
).import_ttls()
````
